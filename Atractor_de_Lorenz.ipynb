## Atractor de Lorenz

$$ \dot{x}=\sigma (y-x) $$
$$ \dot{y}=\rho x -y -xz $$
$$ \dot{z}=-\beta z +xy $$

* x temperatura
* y velocidad del viento
* z la humedad

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
x0,y0,z0=0,1,1.05
sigma,beta,rho=10,2.667,28
tiempo=np.linspace(0,100,10000)
```

```python
sistema= lambda var,t : [sigma*(var[1]-var[0]),rho*var[0]-var[1]-var[0]*var[2],-beta*var[2]+var[0]*var[1]]
```

```python
sol=odeint(sistema,[x0,y0,z0],tiempo)
xx,yy,zz=sol.T
```

### Soluci√≥n X contra Y

```python
plt.scatter(xx,yy,c=xx,cmap='viridis',alpha=0.8,s=3,label="Atractor de Lorenz X contra Y")
plt.xlabel("Variable x")
plt.ylabel("Variable y")
plt.legend()
plt.show()
```

### Soluci√≥n X contra Z

```python
plt.scatter(xx,zz,c=xx,cmap='plasma',alpha=0.8,s=3,label="Atractor de Lorenz X contra Z")
plt.xlabel("Variable x")
plt.ylabel("Variable z")
plt.legend();
plt.show()
```

### Soluci√≥n Y contra Z

```python
plt.scatter(yy,zz,c=xx,cmap='PiYG',alpha=0.8,s=3,label="Atractor de Lorenz Y contra Z")
plt.xlabel("Variable y")
plt.ylabel("Variable z")
plt.legend()
plt.show()
```
